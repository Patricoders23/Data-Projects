# Regresión Lineal 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df = pd.DataFrame()
df['Area'] = [2100, 2600, 3200, 3290, 3900]
df['Precio'] = [4500000, 5300000, 5850000, 6200000, 6700000]
```

```python
plt.scatter(df['Area'], df['Precio'])
```

```python
# X mayúscula para entrenar el modelo / Variable independiente
X=df[['Area']] #Son 2 corchetes porque el dataframe es bidimensional
type(df[['Area']])
```

```python
#'y' minúscula es la Variable Dependiente que vamos a predecir 
y = df['Precio']
```

```python
#Modelo de Regresión Lineal
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X,y) #Entrenamos el modelo, ve las relaciones 
```

```python
modelo.predict([[3500]])
#Para una área de 3500 predice el precio de 6'296_285 
```

```python
plt.plot(df['Area'],
        modelo.predict(X));
```

Ejercicio 2

```python
df = pd.DataFrame({'Edad': [25, 30, 35, 40, 45],
                   'Salario': [3000, 3500, 4000, 4500, 5000]})

plt.scatter(df['Edad'], df['Salario'], color='blue', marker='o')

plt.xlabel('Edad')
plt.ylabel('Salario')
plt.title('Relación entre Edad y Salario')
plt.grid(True)
```

```python
X=df[['Edad']]
y=df['Salario']

modelo = linear_model.LinearRegression()
modelo.fit(X,y)

plt.plot(df['Edad'],
        modelo.predict(X));
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.title('Relación entre Edad y Salario')
plt.grid(True)
```

```python
plt.scatter(df['Edad'], df['Salario'], color='blue', marker='o', label='Datos')
plt.plot(df['Edad'], modelo.predict(X), color='red', label='Regresión Lineal')

plt.xlabel('Edad')
plt.ylabel('Salario')
plt.title('Relación entre Edad y Salario')
plt.grid(True)
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression

# Crear DataFrame
data = {'Edad': [25, 30, 35, 40, 45],
        'Salario': [3000, 3500, 4000, 4500, 5000]}
df = pd.DataFrame(data)

X = df[['Edad']]
y = df['Salario']

# Crear el modelo de regresión lineal
modelo = linear_model.LinearRegression()
modelo.fit(X, y)

# Crear el diagrama de dispersión
plt.scatter(df['Edad'], df['Salario'], color='blue', marker='o')

plt.xlabel('Edad')
plt.ylabel('Salario')
plt.title('Relación entre Edad y Salario')
plt.grid(True)
plt.legend()
plt.plot(df['Edad'], modelo.predict(X));

# Mostrar el gráfico
plt.show()
```
