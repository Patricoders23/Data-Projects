# Clustering Jerárquico (ML No Supervisado) 

El Clustering Jerárquico es un método de análisis de clusters que construye una jerarquía de clusters, lo cual difiere de otros métodos que requieren la especificación previa del número de clusters.

Los dos enfoques principales del Clustering Jerárquico son Aglomerativo (Bottom-up) y Divisivo (Top-down), que manejan la formación de clusters de maneras opuestas.

Un dendrograma es una herramienta visual que proporciona una representación gráfica de cómo se combinan los clusters a diferentes niveles de similitud en el Clustering Jerárquico.


```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt 
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
iris = load_iris()
```

```python
X = iris.data
```

El método de enlace 'ward' utilizado en Clustering Jerárquico minimiza la varianza de los clusters que se van fusionando, asegurando una mayor homogeneidad dentro de los clusters.

```python
#Guardamos el cluster en una variable
linked = linkage(X, 'ward') #ward, minimiza la varianza
```

El parámetro show_leaf_counts=True en la función dendrograma añade números al final de las etiquetas de las hojas para indicar el número de puntos de datos originales en ese cluster, lo que facilita la interpretación de la estructura de clusters.

```python
plt.figure(figsize=(10,7))
dendrogram(linked,  
          orientation='top', #Dibuje las raíces en la parte superior y las hojas hacia abajo
          labels=iris.target,
          distance_sort='descending', 
          show_leaf_counts=True) #Añade números al final de las hojas y entender el tamaño de los clusters
plt.title('Dendograma de Clustering Jerárquico')
plt.xlabel('Índice de la Muestra')
plt.ylabel('Distancia')
```
