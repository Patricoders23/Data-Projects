```python
#Hexbin plot 
plt.style.use('ggplot')

fig, ax = plt.subplots()

hb = ax.hexbin(df['Latitud'], df['Longitud'], C=df['Luminosidad'], 
          gridsize=50, 
          cmap='Greens')
ax.set_facecolor('#000000')
plt.colorbar(hb, label='nivel de sombras')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.title('Luminosidad en Villa Luz');
```

```python
#Stem plot
x=np.arange(1,11)
y=x**2
plt.stem(x,y)
plt.title("Relacin entre Nmeros y sus Cuadrados")
plt.xlabel("Nmero (x)")
plt.ylabel("Cuadrado del nmero (x^2)");
```

```python
plt.style.use('_mpl-gallery')

# make data
x = np.arange(1, 6)
y1 = [6,1,4,4,8]
y2 = [4,6,3,5,8]
y3 = [7,9,9,2,7]
y = np.vstack([y1, y2, y3])

# plot
fig, ax = plt.subplots()

ax.stackplot(x, y);
```

```python
#Stackplot
x = np.arange(1, 6)
y1 = [6,1,4,4,8]
y2 = [4,6,3,5,8]
y3 = [7,9,9,2,7]
y = np.vstack([y1, y2, y3])

# plot
fig, ax = plt.subplots()

ax.stackplot(x, y, 
             labels=['Serie y1', 'Serie y2', 'Serie y3'], 
             colors=['red', 'blue', 'green'])
ax.set_title("Distribucin acumulativa de valores por serie de tiempo")
ax.set_xlabel("Tiempo")
ax.set_ylabel("Valores Acumulados")
ax.legend(loc='upper left');
```

```python
#Violin plot
# Generar los datos para cada muestra usando numpy.random.normal
np.random.seed(0) # Establecer la misma semilla para la generacin de datos
data1 = np.random.normal(loc=0, scale=0.5, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=-2, scale=1.5, size=100)

data = [data1, data2, data3]

# Crear el grfico
fig, ax = plt.subplots()

# Crear el violin plot
vp = ax.violinplot(data)

# Establecer etiquetas en el eje x
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Muestra 1', 'Muestra 2', 'Muestra 3'])

# Agregar ttulo
ax.set_title("Comparacin de distribuciones en Violin Plot");
```
