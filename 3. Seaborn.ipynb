# Seaborn

```python
#Importamos librerías
import seaborn as sns 
import pandas as pd 
import numpy as np 
```

## Relplot (relacional) --> scatterplot, lineplot

## Displot (distribución) --> histplot, kdeplot, ecdfplot, rugplot

-Usada para visualizar la distribución de valores en un conjunto de datos.
-

```python
#Displot representación distributiva 
pinguinos = sns.load_dataset('penguins')
pinguinos
```

```python
sns.displot(data=pinguinos,
           x='body_mass_g');
```

```python
sns.displot(data=pinguinos,
           x='body_mass_g',
           kde=True);#Distribución de la densidad de kernel
```

```python
sns.displot(data=pinguinos,
           x='body_mass_g',
           kde=True,
           col='island');
```

```python
sns.displot(data=pinguinos,
           x='flipper_length_mm',
           kind='kde',
           rug=True);#Util para ver la densidad de los datos en el eje x
#rug=True añade marcas individuales para cada observación en el gráfico, lo que ayuda a entender mejor la densidad y distribución de los datos.
```

```python
sns.displot(data=pinguinos,
           x='flipper_length_mm',
           kind='ecdf');#Agregan los datos y obtenemos la totalidad 
#ECDF muestra la proporción de observaciones que caen por debajo de cada valor específico en el conjunto de datos.
```

## Catplot (categórica) -->stripplot, swarmplot, boxplot, violinplot, pointplot, barplot

-Representa categorías de variables en gráficos

```python
sns.catplot(data=propinas,
           kind='strip',
           x='day',
           y='total_bill',
           hue='smoker');
```

```python
sns.catplot(data=propinas,
           kind='swarm', #Esparce los puntos
           x='day',
           y='total_bill',
           hue='smoker');
```

```python
sns.catplot(data=propinas,
           kind='box',
           x='day',
           y='total_bill',
           hue='smoker');
#En los boxplot, la caja central representa el rango intercuartílico, que es la diferencia entre el cuartil 25 y el cuartil 75.
```

```python
sns.catplot(data=propinas,
           kind='violin',
           x='day',
           y='total_bill',
           hue='smoker');
#Los violinplot incorporan una representación de la densidad de los datos, lo que proporciona una visión más detallada de la distribución que un boxplot
```

```python
sns.catplot(data=propinas,
           kind='point',#
           x='day',
           y='total_bill',
           hue='smoker');
#Solo el viernes el promedio de no-fumadores supera a los fumadores
```

Los barplot calculan y muestran pequeñas líneas negras en la parte superior de cada barra que representan los intervalos de confianza, brindando una idea de la incertidumbre estadística de las estimaciones.

```python
sns.catplot(data=propinas,
           kind='bar',#Líneas representa intervalos de confianza (menor tamaño --> más confianza)
           x='day',
           y='total_bill',
           hue='smoker');
```

## Jointplot

jointplot() es específicamente útil para analizar la relación entre dos variables numéricas en un dataset, mostrando tanto la distribución conjunta en el centro como las distribuciones marginales.

```python
## Joinplot --> relación entre características 
sns.jointplot(data=pinguinos,
            x='flipper_length_mm',
            y='bill_length_mm',
            hue='species');
```

## Pairplot

pairplot() es beneficioso porque automáticamente toma todas las variables numéricas de un dataset y crea un grid de gráficos para cada par de variables, lo que permite una evaluación rápida y comprensiva de todas las relaciones posibles.

Utilizar la opción hue en pairplot() permite colorear los puntos en cada gráfico del grid según los valores de una variable categórica, facilitando la identificación de patrones o diferencias entre categorías.

```python
sns.pairplot(data=pinguinos,
            hue='species');
```

## Funciones de Nivel Inferior 

Las funciones de nivel inferior ofrecen un control más detallado sobre la estética y disposición de los gráficos, a diferencia de las funciones de alto nivel que son más rápidas y sencillas.

### PairGrid
PairGrid() es utilizado para mostrar la relación entre pares de variables, permitiendo visualizar cómo se relacionan entre sí en distintos gráficos.

*El atributo hue se utiliza para especificar la variable que determinará los colores de los gráficos, ayudando a diferenciar grupos dentro de los datos.

```python
g = sns.PairGrid(pinguinos,
                hue='species');
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True);
#El atributo corner=True permite que solo se muestren los gráficos de la esquina inferior izquierda, eliminando gráficos duplicados que representan las mismas relaciones con los ejes invertidos
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True);
g.map_lower(sns.kdeplot,
           hue=None,
           level=5,
           color='.2');
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True);
g.map_lower(sns.kdeplot,
           hue=None,
           level=5,
           color='.2');
g.map_lower(sns.scatterplot,
           marker='+');
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True);
g.map_lower(sns.kdeplot,
           hue=None,
           level=5,
           color='.2');
g.map_lower(sns.scatterplot,
           marker='+');
g.map_diag(sns.histplot,
          element='step',
          linewidth=0,
          kde=True)
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True);
g.map_lower(sns.kdeplot,
           hue=None,
           level=5,
           color='.2');
g.map_lower(sns.scatterplot,
           marker='+');
g.map_diag(sns.histplot,
          element='step',
          linewidth=0,
          kde=True)
g.add_legend(frameon=True);
```

```python
g = sns.PairGrid(pinguinos,
                hue='species',
                corner=True);
g.map_lower(sns.kdeplot,
           hue=None,
           level=5,
           color='.2');
g.map_lower(sns.scatterplot,
           marker='+');
g.map_diag(sns.histplot,
          element='step',
          linewidth=0,
          kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61,.6));
```

## Nivel de figuras vs. Nivel de Axes en Seaborn 

--> ´histplot()´ en Seaborn es un ejemplo de una función que opera a nivel del axes, es decir, dibuja los datos directamente en un objeto de matplotlib, a diferencia de ´relplot()´, ´displot()´ y ´catplot()´ que operan a nivel de la figura.

--> Las funciones a nivel de figura como relplot(), displot(), y catplot() en Seaborn devuelven un objeto FacetGrid, que permite manipular y configurar los gráficos compuestos de manera más flexible.

--> Un objeto FacetGrid ofrece métodos y propiedades que facilitan la personalización y manipulación de los gráficos a un nivel más detallado, como ajustar etiquetas, añadir títulos, entre otros.

--> A nivel de figura, algunas funciones pueden no permitir el acceso a todos los parámetros específicos de los gráficos de nivel del axes, lo que puede resultar en una pérdida de flexibilidad en la personalización del gráfico.

--> relplot() es una función a nivel de la figura que gestiona su propia figura completa, lo que significa que no está diseñada para ser incorporada en gráficos ya existentes. A diferencia de scatterplot(), histplot() y stripplot(), que son más flexibles en este aspecto al operar a nivel del axes.

```python
sns.displot(data=pinguinos, 
            x='flipper_length_mm',
            hue='species',
            multiple='stack');#Apilan las especies 
```

```python
sns.histplot(data=pinguinos, 
            x='flipper_length_mm',
            hue='species',
            multiple='stack');
```

```python
sns.displot(data=pinguinos, 
            x='flipper_length_mm',
            hue='species',
            multiple='stack',
            col='species');
```
